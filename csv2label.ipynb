```python
import os
import shutil
```

```python
workspace = 'out2.csv'
destname = 'csv_label2'
# shutil.rmtree(destname)
```

```python
def csv_label(workspace):
   with open(workspace) as f:
       for line in f.readlines():
            
            label = line.strip().split('..')[-1]
            path = line.strip().split('..')[-2].split('/')[-1]
            
            dstfile = os.path.join(destname,label)
            print(dstfile)
            if not os.path.exists(dstfile):
                os.makedirs(dstfile)
            srcfile = os.path.join(line.strip().split('..')[-2].split('/')[1],path)
            print(srcfile)
            shutil.move(srcfile,dstfile)
csv_label(workspace)        
```
